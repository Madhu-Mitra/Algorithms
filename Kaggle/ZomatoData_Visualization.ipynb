```python
# import the libraries

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# Input data files are available in the "/resources" directory.
```

```python
Zomato_filepath = "resources/zomato.csv"
restaurant_data=pd.read_csv(Zomato_filepath)
del restaurant_data['url']
del restaurant_data['address']
del restaurant_data['phone']
```

```python
#fill all null values with NaN
restaurant_data = restaurant_data.fillna(np.nan)
restaurant_data.sample(n=1)
```

```python
#encoding the lables so that they are all of same data type
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
restaurant_data.iloc[:,1] = labelencoder.fit_transform(restaurant_data.iloc[:,1])
restaurant_data.iloc[:,2] = labelencoder.fit_transform(restaurant_data.iloc[:,2])
restaurant_bookingdata=restaurant_data[['name' , 'online_order','book_table']]
restaurant_data.head()
```

```python
plt.figure(figsize=(20,8))
plot=sns.countplot(x="rest_type",data=restaurant_data, palette = "Set1")
plot.set_xticklabels(plot.get_xticklabels(), rotation=45, ha="right")
plt.show()
#below bar graph is displayed. it clearly depicts what kind of dining is more popular
```

```python
slices=restaurant_data['online_order'].value_counts().sort_values(ascending=True)
slices
```
